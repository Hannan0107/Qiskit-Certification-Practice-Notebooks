# Quantum Volume in Qiskit

---

# Definition

Quantum Volume (QV) is a benchmarking metric used to measure the overall capability of a quantum computer.

Introduced by IBM.

---

# Main Idea

Quantum Volume measures how well a quantum computer can execute difficult random quantum circuits.

It evaluates:

- qubit quality
- gate fidelity
- connectivity
- noise
- circuit depth
- compiler efficiency

---

# Important Concept

Quantum Volume does NOT only measure:

$$
\text{number of qubits}
$$

It measures:

$$
\text{usable computational capability}
$$

---

# QuantumVolume Circuit

In Qiskit:

```python
from qiskit.circuit.library import QuantumVolume

qc = QuantumVolume(4,4)

qc.draw('mpl')
```

---

# Parameters

```python
QuantumVolume(num_qubits, depth)
```

Example:

```python
QuantumVolume(4,4)
```

means:

| Parameter | Meaning |
|---|---|
| 4 qubits | circuit width |
| depth 4 | number of random layers |

---

# What the Circuit Contains

Quantum Volume circuits use:

- random two-qubit gates
- random entanglement
- layered random operations

These circuits are intentionally difficult for hardware.

---

# Why Random Circuits?

Random circuits stress-test:

- gate accuracy
- decoherence
- connectivity
- transpilation quality

---

# What Quantum Volume Measures

| Feature | Tested |
|---|---|
| Qubit quality | Yes |
| Noise resistance | Yes |
| Gate fidelity | Yes |
| Circuit depth survival | Yes |
| Connectivity | Yes |

---

# Important Formula

Quantum Volume is often written as:

$$
QV = 2^k
$$

Where:

$$
k
$$

is the largest successful square random circuit size.

---

# Example

If hardware successfully executes:

- 8-qubit
- depth-8

random circuits,

then:

$$
QV = 2^8 = 256
$$

---

# Important Distinction

| Metric | Meaning |
|---|---|
| Qubit count | hardware size only |
| Quantum Volume | practical capability |

---

# Why Quantum Volume Matters

A quantum computer with:

- many noisy qubits

may perform worse than:

- fewer high-quality qubits

Quantum Volume captures this tradeoff.

---

# Benchmark Analogy

Quantum Volume is like a stress-test benchmark.

Example:

| Classical System | Benchmark |
|---|---|
| CPU | benchmark program |
| GPU | rendering benchmark |
| Quantum computer | Quantum Volume circuit |

---

# Important Concept

Quantum Volume circuits are NOT useful algorithms themselves.

They are benchmarking circuits used to evaluate hardware performance.

---

# Applications

Used for:

- hardware benchmarking
- comparing quantum processors
- measuring hardware improvements
- evaluating NISQ systems

---

# Core Idea

Quantum Volume in Qiskit generates randomized benchmark circuits used to measure the practical computational capability and reliability of quantum hardware.

````python
from qiskit.circuit.library import PhaseEstimation
from qiskit.circuit.library import RXGate
import numpy as np

num_eval_qubits = 3

unitary = RXGate(np.pi/2)

phase_est = PhaseEstimation(num_eval_qubits, unitary)

phase_est.draw('mpl')
````

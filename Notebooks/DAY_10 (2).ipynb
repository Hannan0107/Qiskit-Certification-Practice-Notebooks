# N-Local Circuits in Qiskit

N-Local circuits are parameterized quantum circuits used as ansatzes in variational quantum algorithms such as:

- VQE (Variational Quantum Eigensolver)
- QAOA (Quantum Approximate Optimization Algorithm)
- Quantum Machine Learning

---

# Main Idea

N-local circuits are built using repeating layers of:

1. Rotation gates
2. Entanglement gates

---

# General Structure

```text
Rotation Layer
      ↓
Entanglement Layer
      ↓
Rotation Layer
      ↓
Entanglement Layer
```

Repeated multiple times.

---

# Why Called “N-Local”?

Because gates interact locally between subsets of:

$$
n
$$

qubits.

Usually:

- neighboring qubits
- pairwise entanglement

---

# Important Concept

These circuits are parameterized.

Example:

$$
R_Y(\theta)
$$

The parameters:

$$
\theta
$$

are optimized during training.

---

# Used In Variational Algorithms

Variational algorithms work like:

1. Create parameterized circuit
2. Run quantum circuit
3. Measure cost
4. Optimize parameters classically

---

# Data Preparation Circuits in Qiskit

Data preparation circuits initialize qubits into desired quantum states before computation begins.

Used in:

- quantum algorithms
- quantum machine learning
- amplitude encoding
- state initialization
- simulation

---

# Main Idea

By default, qubits start in:

$$
|0\rangle
$$

Data preparation circuits transform them into custom quantum states.

---

# StatePreparation

## Complex Amplitude State Preparation

Prepares arbitrary quantum states with custom amplitudes.

General state:

$$
|\psi\rangle
=
\alpha_0|0\rangle
+
\alpha_1|1\rangle
+
\alpha_2|2\rangle
+\cdots
$$

Where amplitudes may be complex numbers.

---

# Example

Prepare state:

$$
|\psi\rangle
=
\frac{|0\rangle+|1\rangle}{\sqrt2}
$$

using:

```python
StatePreparation([1/sqrt(2), 1/sqrt(2)])
```

---

# Important Concept

The amplitudes must satisfy normalization:

$$
\sum_i |\alpha_i|^2 = 1
$$

because total probability must equal 1.

---

# Purpose

Used for:

- amplitude encoding
- custom quantum states
- quantum ML
- algorithm initialization

```python
StatePreparation(statevector)
```

---

# What It Does Internally

Qiskit decomposes the desired state into:

- rotations
- entangling gates
- phase operations

to physically create the target quantum state.

---

# Initialize

## Complex Amplitude Initialization

Initializes qubits directly into a chosen state.

Similar to:

```python
StatePreparation
```

but includes reset operations.

---

# Important Difference

`Initialize` is:

```text
Instruction
```

NOT:

```text
Gate
```

---

# Why?

Because it internally uses:

$$
\text{reset}
$$

operations.

---

# Reset Operation

Reset forces a qubit into:

$$
|0\rangle
$$

regardless of previous state.

This operation is:

- non-unitary
- irreversible

---

# Important Quantum Rule

Quantum gates must be:

- unitary
- reversible

Since reset is NOT unitary:

```text
Initialize
```

cannot be classified as a true quantum gate.

---

# Example

```python
qc.initialize([1/sqrt(2),1/sqrt(2)],0)
```

Initializes qubit into:

$$
\frac{|0\rangle+|1\rangle}{\sqrt2}
$$

---

# Difference Between StatePreparation and Initialize

| Feature | StatePreparation | Initialize |
|---|---|---|
| Type | Gate | Instruction |
| Uses reset | No | Yes |
| Unitary | Yes | No |
| Reversible | Yes | No |

---

# Important Concept

---

# StatePreparation

Creates the target state using only reversible quantum operations.

---

# Initialize

First resets qubits, then prepares the state.

So it can overwrite existing quantum states.

---

# Preparing a Specific Quantum State in Qiskit

Goal:

Prepare the Bell state:

$$
|\psi\rangle
=
\frac{|00\rangle+|11\rangle}{\sqrt2}
$$

This is an entangled 2-qubit quantum state.

---

# Mathematical Meaning

The state:

$$
\frac{|00\rangle+|11\rangle}{\sqrt2}
$$

means:

| Basis State | Amplitude |
|---|---|
| \(|00\rangle\) | \(1/\sqrt2\) |
| \(|01\rangle\) | 0 |
| \(|10\rangle\) | 0 |
| \(|11\rangle\) | \(1/\sqrt2\) |

---

# Important Concept

For 2 qubits there are:

$$
2^2 = 4
$$

basis states:

$$
|00\rangle,\ |01\rangle,\ |10\rangle,\ |11\rangle
$$

So the amplitude vector must contain 4 values.

---

# State Vector

The amplitudes are written as:

$$
\left[
\frac1{\sqrt2},
0,
0,
\frac1{\sqrt2}
\right]
$$

Corresponding to:

| Index | Basis State |
|---|---|
| 0 | \(|00\rangle\) |
| 1 | \(|01\rangle\) |
| 2 | \(|10\rangle\) |
| 3 | \(|11\rangle\) |

---

# Qiskit Code

```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import StatePreparation
import numpy as np

# Create 2-qubit circuit
qc = QuantumCircuit(2)

# Define amplitudes for:
# |00>, |01>, |10>, |11>
state = [
    1/np.sqrt(2),
    0,
    0,
    1/np.sqrt(2)
]

# Prepare the quantum state
prep = StatePreparation(state)

qc.append(prep, [0,1])

qc.draw('mpl')
```

---

# Why 4 Amplitudes?

Because:

$$
2^n
$$

amplitudes are needed for:

$$
n
$$

qubits.

For:

| Qubits | Basis States |
|---|---|
| 1 | 2 |
| 2 | 4 |
| 3 | 8 |

---

# Normalization Check

Quantum states must satisfy:

$$
\sum_i |\alpha_i|^2 = 1
$$

Here:

$$
\left(\frac1{\sqrt2}\right)^2
+
\left(\frac1{\sqrt2}\right)^2
=
\frac12+\frac12
=
1
$$

So the state is valid.

---

# Physical Interpretation

The system is in superposition of:

$$
|00\rangle
\quad\text{and}\quad
|11\rangle
$$

only.

---

# Measurement Outcomes

Measurement probabilities:

| State | Probability |
|---|---|
| 00 | 50% |
| 11 | 50% |
| 01 | 0% |
| 10 | 0% |

---

# Important Observation

This is an entangled Bell state.

The qubits are correlated:

- if first qubit is 0 → second is also 0
- if first qubit is 1 → second is also 1

---

# Alternative Way (Using Gates)

Same state can also be created with:

```python
qc.h(0)
qc.cx(0,1)
```

because:

1. Hadamard creates superposition
2. CNOT entangles qubits

---

# Core Idea

State preparation circuits encode desired amplitudes into qubits to create specific quantum states such as Bell states and arbitrary superpositions.

````python
from qiskit.circuit.library import StatePreparation
from qiskit import QuantumCircuit
import numpy as np

Target_State = [1/np.sqrt(2), 0, 0, 1/np.sqrt(2)]

StatPrep = StatePreparation(Target_State)
print(StatPrep.num_qubits)
qc = QuantumCircuit(StatPrep.num_qubits)
qc.append(StatPrep, range(StatPrep.num_qubits))
qc.draw('mpl')
````

````python
qc.decompose().draw('mpl')
````

````python
qc.decompose().decompose().draw('mpl')
````

````python
qc.decompose().decompose().decompose().draw('mpl')
````

````python
qc.decompose().decompose().decompose().decompose().draw('mpl')
````

# Oracles in Qiskit

Oracles are black-box quantum operations used inside many quantum algorithms.

Used heavily in:

- Grover’s algorithm
- Bernstein-Vazirani
- Deutsch-Jozsa
- Simon’s algorithm
- search and optimization problems

---

# Main Idea

An oracle encodes a hidden function:

$$
f(x)
$$

inside a quantum circuit.

The algorithm can query the oracle without directly knowing how it works internally.

---

# Why Called “Black Box”?

Because the algorithm treats the oracle like:

```text
input → hidden computation → output
```

without needing to know the implementation details.

---

# Classical Analogy

Suppose a machine:

- receives input
- secretly computes something
- returns answer

You can use it without knowing its internal logic.

That is a black-box oracle.

---

# Quantum Oracle

Quantum oracles encode Boolean functions into quantum operations.

Example:

$$
f(x)\in\{0,1\}
$$

---

# Types of Oracle Encodings

| Type | Behavior |
|---|---|
| Phase Oracle | modifies phase |
| Bit-Flip Oracle | flips output qubit |

---

# Gate Representation vs QuantumCircuit Representation

| Type | Meaning |
|---|---|
| Gate Representation | reusable oracle gate |
| QuantumCircuit Representation | full oracle circuit |

---

# PhaseOracleGate

## Phase Oracle Gate

Implements a phase oracle.

Instead of changing bits, it changes phase.

---

# Main Action

If:

$$
f(x)=1
$$

then:

$$
|x\rangle
\rightarrow
- |x\rangle
$$

Otherwise:

$$
|x\rangle
\rightarrow
|x\rangle
$$

---

# Mathematical Form

:contentReference[oaicite:0]{index=0}

---

# Example

Suppose:

$$
f(10)=1
$$

Then:

$$
|10\rangle
\rightarrow
-|10\rangle
$$

Phase changes but measurement probabilities initially remain same.

---

# Important

Quantum interference later uses these phase flips to amplify correct answers.

---

# Used In

- Grover search
- amplitude amplification
- optimization algorithms

```python
PhaseOracleGate(...)
```

---

# BitFlipOracleGate

## Bit-Flip Oracle Gate

Implements oracle by flipping a target qubit.

---

# Mathematical Action

:contentReference[oaicite:1]{index=1}

Where:

| Symbol | Meaning |
|---|---|
| \(x\) | input register |
| \(y\) | target/output qubit |
| \(\oplus\) | XOR |

---

# Example

If:

$$
f(x)=1
$$

then:

$$
y
\rightarrow
y\oplus1
$$

which flips the target qubit.

---

# Example

Input:

$$
|x,0\rangle
$$

becomes:

$$
|x,1\rangle
$$

if:

$$
f(x)=1
$$

---

# Difference From Phase Oracle

| Phase Oracle | BitFlip Oracle |
|---|---|
| changes phase | flips qubit |
| no target qubit needed | target qubit required |
| used in Grover | general Boolean computation |

---

# PhaseOracle

## QuantumCircuit Representation

Full circuit implementation of a phase oracle.

Implements Boolean expressions as phase-flip quantum circuits.

---

# Example Boolean Expression

```python
PhaseOracle("(a & b) | c")
```

Creates oracle for:

$$
(a \land b)\lor c
$$

---

# Important Concept

The oracle marks “correct” solutions by changing their phase.

Grover’s algorithm then amplifies those marked states.

---

# Oracle in Grover Search

Grover workflow:

```text
Superposition
      ↓
Oracle marks solution
      ↓
Diffusion amplifies solution
      ↓
Measurement
```

---

# Why Phase Flip Matters

Phase itself is not directly measurable.

But interference between amplitudes causes:

- correct states to grow
- incorrect states to shrink

---

# Example

Suppose:

$$
|00\rangle,|01\rangle,|10\rangle,|11\rangle
$$

If oracle marks:

$$
|10\rangle
$$

then:

$$
|10\rangle \rightarrow -|10\rangle
$$

Later interference amplifies that state.

---

# Important Quantum Idea

The oracle does NOT directly “output the answer.”

It modifies amplitudes/phases so the quantum algorithm can discover the answer using interference.

---

# Summary Table

| Component | Purpose |
|---|---|
| PhaseOracleGate | phase-flip oracle |
| BitFlipOracleGate | target-bit flipping oracle |
| PhaseOracle | full phase-oracle circuit |

---

# Core Idea

Oracles are black-box quantum operations that encode hidden Boolean functions into quantum circuits using phase flips or controlled bit flips for use in quantum algorithms like Grover search.

## Boolean Expression

$
(a \land b) \lor c
$

Meaning:

- $a \land b$ : both $a=1$ AND $b=1$
- $\lor c$ : OR $c=1$

---

## Truth Table

| a | b | c | Result |
|---|---|---|---|
| 0 | 0 | 0 | False |
| 0 | 0 | 1 | True |
| 0 | 1 | 0 | False |
| 0 | 1 | 1 | True |
| 1 | 0 | 0 | False |
| 1 | 0 | 1 | True |
| 1 | 1 | 0 | True |
| 1 | 1 | 1 | True |

---

## Correct (Marked) States

The Boolean expression evaluates to TRUE for:

$
001,\ 011,\ 101,\ 110,\ 111
$

````python
from qiskit.circuit.library import BitFlipOracleGate

Expression = "(a | b | c)"
Orac = BitFlipOracleGate(Expression)
qc = QuantumCircuit(Orac.num_qubits)
qc.append(Orac, range(Orac.num_qubits))
qc.draw('mpl')
````

````python
qc.decompose().draw('mpl')
````

# Template Circuits in Qiskit

Template circuits are predefined quantum circuits that compute the **identity operation**.

This means:

$
U = I
$

where:

- \(U\) = quantum operation
- \(I\) = identity matrix

So overall, the circuit does **nothing** to the quantum state.

---

# Why are Templates Important?

Templates are mainly used for:

- Circuit optimization
- Gate reduction
- Reducing circuit depth
- Hardware-efficient transpilation

The Qiskit transpiler uses templates to replace inefficient gate patterns with better ones.

---

# Main Idea

Two different gate sequences can perform the same operation.

Example:

$
H Z H = X
$

So instead of using 3 gates:

```text
H → Z → H
```

the transpiler may replace it with:

```text
X
```

This makes the circuit smaller and faster.

---

# Types of Template Circuits

## 1. NCT Templates

NCT stands for:

- N = NOT gate
- C = Controlled-NOT gate
- T = Toffoli gate

Uses:

- XGate
- CXGate
- CCXGate

These templates are commonly used in:

- Reversible computing
- Boolean logic synthesis
- Grover oracles

---

## 2. Clifford Templates

Uses Clifford gates such as:

- H
- S
- X
- Z
- CX

Clifford circuits are important because they:

- are easy to simulate classically
- are used in quantum error correction
- are used in stabilizer circuits

Example:

$
H Z H = X
$

---

## 3. RZX Templates

Uses:

$
RZX(\theta)
$

gate.

RZX templates are important for:

- IBM quantum hardware optimization
- Pulse-efficient transpilation
- Native hardware gate decomposition

These templates help generate circuits better suited for real quantum devices.

---

# Identity Property

Since templates compute identity:

$
A \cdot I \cdot B = A \cdot B
$

the transpiler can safely:

- insert templates
- remove templates
- replace gate sequences

without changing the final computation.

---

# Simple Intuition

Template circuits are like algebraic identities.

Example in mathematics:

$
x^2 - y^2 = (x-y)(x+y)
$

Both expressions are equivalent.

Similarly in quantum computing:

```text
Different gate arrangements
Same quantum operation
```

The transpiler chooses the more efficient form.

````python
from qiskit.circuit.library.templates import template_nct_4b_1

Template = template_nct_4b_1()
print(Template.num_qubits)
````

````python
identity = np.identity(2 ** len(Template.qubits), dtype=complex)
print(identity)
````

````python
from qiskit.quantum_info import Operator

data = Operator(Template).data
print(data)
````

````python
is_identity = np.allclose(data, identity)
print(is_identity)
````

````python
Template.draw('mpl')
````

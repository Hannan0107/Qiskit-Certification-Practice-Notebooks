# Standard Gates and Operations in Qiskit

## 1. Reversible and Unitary Quantum Gates

Quantum gates in Qiskit are generally:

- **Unitary**
- **Reversible**

A quantum gate is represented by a unitary matrix $U$.

### Unitary Condition

$$
U^\dagger U = I
$$

Where:

- $U^\dagger$ = conjugate transpose
- $I$ = identity matrix

This ensures:

- probabilities are preserved
- quantum states remain valid

---

## 2. Reversible Operations

Quantum operations can always be undone.

### Example: X Gate

$$
X =
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$$

Applying X twice:

$$
X^2 = I
$$

Meaning:

$$
|0\rangle \rightarrow |1\rangle \rightarrow |0\rangle
$$

So the X gate is reversible.

---

# 3. Abstract Operations

Some Qiskit operations describe:

$$
\text{WHAT the operation does}
$$

rather than:

$$
\text{HOW the hardware implements it}
$$

Example:

```python
from qiskit.circuit.library import MCXGate
```

MCX = Multi-Controlled X gate.

---

## 4. MCXGate

MCX flips the target qubit only when all control qubits are $1$.

Example:

```python
MCXGate(3)
```

Means:

- 3 control qubits
- 1 target qubit

Logical action:

$$
|1110\rangle \rightarrow |1111\rangle
$$

---

## 5. Different Decompositions

MCX does not have one fixed hardware implementation.

Qiskit may decompose it differently depending on:

- backend hardware
- optimization level
- available ancilla qubits
- basis gates

Example decomposition:

$$
H \rightarrow CX \rightarrow T \rightarrow T^\dagger \rightarrow CX
$$

Different decompositions can produce the same logical operation.

---


$$

It usually contains:

- rotation layers
- entanglement layers

Defines mathematical behavior.

Can have multiple decompositions.

---

### Structural Operation

Example:

$$

Defines a fixed circuit pattern.

Has a unique structure.

# Quantum Amplitudes are "Spread Out"

In quantum computing, gates do not simply flip values like classical logic.

Instead, they distribute the quantum state into multiple amplitudes.

---

# Key Idea

A qubit state is written as:

$$
|\psi\rangle = a|0\rangle + b|1\rangle
$$

Where:

- $a$ = amplitude of $|0\rangle$
- $b$ = amplitude of $|1\rangle$

These amplitudes are usually complex numbers.

---

# Important Rule

Quantum states preserve:

$$
|a|^2 + |b|^2 = 1
$$

NOT:

$$
a+b=1
$$

This is one of the most important ideas in quantum mechanics.

---

# Example 1 — Hadamard Gate

Applying Hadamard to $|0\rangle$:

$$
|0\rangle
\rightarrow
\frac{1}{\sqrt2}|0\rangle
+
\frac{1}{\sqrt2}|1\rangle
$$

Numerically:

$$
0.707|0\rangle + 0.707|1\rangle
$$

---

## Check the Probabilities

Probability of measuring $|0\rangle$:

$$
|0.707|^2 = 0.5
$$

Probability of measuring $|1\rangle$:

$$
|0.707|^2 = 0.5
$$

Total probability:

$$
0.5 + 0.5 = 1
$$

---

# Example 2 — Square Root of X Gate

The $\sqrt{X}$ gate partially flips the qubit.

Matrix:

$$
\sqrt{X}=
\begin{bmatrix}
0.5+0.5i & 0.5-0.5i \\
0.5-0.5i & 0.5+0.5i
\end{bmatrix}
$$

Applying it once creates a partial rotation.

Applying it twice gives the full X gate:

$$
(\sqrt{X})^2 = X
$$

---

# Example 3 — Adding Complex Amplitudes

Suppose amplitudes are:

$$
0.14644661-0.35355339i
$$

and

$$
0.85355339+0.35355339i
$$

Add real parts:

$$
0.14644661 + 0.85355339 = 1
$$

Add imaginary parts:

$$
-0.35355339i + 0.35355339i = 0
$$

Final result:

$$
1 + 0i = 1
$$

The imaginary parts cancel out.

---

# Intuition

Quantum gates spread the state across amplitudes.

Instead of:

$$
|0\rangle \rightarrow |1\rangle
$$

you often get:

$$
|0\rangle
\rightarrow
a|0\rangle + b|1\rangle
$$

This is called:

- superposition
- amplitude distribution

---

# Core Concept

Quantum computation preserves:

$$
|a|^2 + |b|^2 = 1
$$

which represents total probability.

# Hadamard Gate Matrix Explanation

---

# Hadamard Gate

The Hadamard gate creates superposition.

Matrix:

$$
H =
\frac{1}{\sqrt2}
\begin{bmatrix}
1 & 1 \\
1 & -1
\end{bmatrix}
$$

Numerically:

$$
\frac{1}{\sqrt2} \approx 0.70710678
$$

So Qiskit prints:

$$
\begin{bmatrix}
0.70710678+0.j & 0.70710678+0.j \\
0.70710678+0.j & -0.70710678+0.j
\end{bmatrix}
$$

---

# What Does `+0.j` Mean?

Qiskit stores quantum matrices as complex numbers.

Example:

```python
0.70710678 + 0.j
```

means:

$$
0.70710678 + 0i
$$

So the imaginary part is simply zero.

---

# Purpose of Hadamard Gate

The Hadamard gate transforms basis states into superposition states.

---

## Transformation of $|0\rangle$

$$
|0\rangle
\rightarrow
\frac{|0\rangle + |1\rangle}{\sqrt2}
$$

Vector form:

$$
\frac1{\sqrt2}
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

---

## Transformation of $|1\rangle$

$$
|1\rangle
\rightarrow
\frac{|0\rangle - |1\rangle}{\sqrt2}
$$

Vector form:

$$
\frac1{\sqrt2}
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
$$

---

# Basis States

$$
|0\rangle =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

$$
|1\rangle =
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

---

# Important Rule

The columns of a quantum gate matrix represent:

$$
\text{Transformed basis states}
$$

---

# First Column

The first column is:

$$
H|0\rangle
=
\frac1{\sqrt2}
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

which becomes:

$$
\begin{bmatrix}
0.707 \\
0.707
\end{bmatrix}
$$

---

# Second Column

The second column is:

$$
H|1\rangle
=
\frac1{\sqrt2}
\begin{bmatrix}
1 \\
-1
\end{bmatrix}
$$

which becomes:

$$
\begin{bmatrix}
0.707 \\
-0.707
\end{bmatrix}
$$

---

# Final Hadamard Matrix

Combining both columns:

$$
H=
\begin{bmatrix}
0.707 & 0.707 \\
0.707 & -0.707
\end{bmatrix}
$$

---

# Why Divide by $\sqrt2$?

Without normalization:

$$
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

has magnitude:

$$
\sqrt{1^2 + 1^2}
=
\sqrt2
$$

Quantum states must have total probability equal to 1.

So divide by $\sqrt2$:

$$
\left(\frac1{\sqrt2}\right)^2
+
\left(\frac1{\sqrt2}\right)^2
=
0.5 + 0.5
=
1
$$

---

# Core Concept

The Hadamard matrix is designed to:

- create equal superposition
- preserve normalization
- keep states orthogonal
- remain unitary


Quantum computing uses **complex numbers** to represent:

- phase
- interference
- rotations
- wave behavior

---

# Complex Number Form

Python uses:

$$
a + bj
$$

where:

$$
j = \sqrt{-1}
$$

Example:

```python
0.707 + 0.j
```

means:

$$
0.707 + 0i
$$

---

# Why Qiskit Shows `+0.j`

Qiskit stores quantum matrices using complex datatype.

Even real numbers are stored as:

```python
0.707 + 0.j
```

for consistency with quantum operations.

---

# Important Concept

Quantum states use:

$$
\textbf{complex amplitudes}
$$

NOT ordinary probabilities.

Probabilities are calculated using:

$$
P = |a|^2
$$

---

# Phase Information

These states are different:

$$
|0\rangle + |1\rangle
$$

$$
|0\rangle - |1\rangle
$$

The minus sign changes the phase.

Phase affects interference and computation.

---

# Euler Formula

Complex numbers describe quantum rotations:

$$
e^{i\theta} = \cos(\theta) + i\sin(\theta)
$$

Used in:

- rotation gates
- phase gates
- Bloch sphere rotations

---

# Core Idea

Complex numbers allow quantum systems to:

- store phase
- create interference
- behave like waves
- perform quantum rotations
- remain unitary

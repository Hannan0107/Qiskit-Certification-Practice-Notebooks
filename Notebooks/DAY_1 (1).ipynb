# Pauli Operators

Pauli operators (Pauli matrices) are fundamental $2 \times 2$ matrices used in quantum mechanics and quantum computing. They act on a single qubit.

---

## 1. Pauli-X (Bit Flip)

$$
X =
\begin{bmatrix}
0 & 1 \\
1 & 0
\end{bmatrix}
$$

- Flips $|0\rangle \leftrightarrow |1\rangle$  
- $X|0\rangle = |1\rangle,\quad X|1\rangle = |0\rangle$

---

## 2. Pauli-Y (Bit + Phase Flip)

$$
Y =
\begin{bmatrix}
0 & -i \\
i & 0
\end{bmatrix}
$$

- Combines bit flip and phase change  
- $Y|0\rangle = i|1\rangle,\quad Y|1\rangle = -i|0\rangle$

---

## 3. Pauli-Z (Phase Flip)

$$
Z =
\begin{bmatrix}
1 & 0 \\
0 & -1
\end{bmatrix}
$$

- Leaves $|0\rangle$ unchanged  
- Flips phase of $|1\rangle$  
- $Z|0\rangle = |0\rangle,\quad Z|1\rangle = -|1\rangle$

---

## 4. Identity Operator

$$
I =
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
$$

- Does nothing to the qubit

---

## Key Properties

### Hermitian and Unitary

$$
X^\dagger = X,\quad Y^\dagger = Y,\quad Z^\dagger = Z
$$

$$
X^2 = Y^2 = Z^2 = I
$$

---

### Non-Commutativity

$$
XY = iZ,\quad YZ = iX,\quad ZX = iY
$$

$$
YX = -iZ,\quad ZY = -iX,\quad XZ = -iY
$$

---

### Basis Representation

Any single-qubit operator:

$$
A = a_0 I + a_1 X + a_2 Y + a_3 Z
$$

---

## Action on a General Qubit

Let:

$$
|\psi\rangle = a|0\rangle + b|1\rangle
$$

Then:

$$
X|\psi\rangle = a|1\rangle + b|0\rangle
$$

$$
Y|\psi\rangle = ai|1\rangle - bi|0\rangle
$$

$$
Z|\psi\rangle = a|0\rangle - b|1\rangle
$$

---

## Geometric Interpretation

- $X$ → rotation about x-axis  
- $Y$ → rotation about y-axis  
- $Z$ → rotation about z-axis  

They correspond to $180^\circ$ rotations on the Bloch sphere.

---

## Applications

- Quantum gates  
- Hamiltonians  
- Quantum error correction  
- Pauli decomposition  
- Quantum machine learning

# Pauli Operators in Qiskit — Notes

## 🔹 What are these?

Given:

- p = Pauli('XZ')
- p2 = Pauli('iYZ')
- p3 = Pauli('-iZX')

These represent **multi-qubit Pauli operators (Pauli strings)**.

---

## 🔹 Meaning of Each Operator

### 1. p = 'XZ'

$$
p = X \otimes Z
$$

- Qubit 1 → X (bit flip)
- Qubit 2 → Z (phase flip)

---

### 2. p2 = 'iYZ'

$$
p2 = i \cdot (Y \otimes Z)
$$

- Y on first qubit, Z on second
- Multiplied by **i (imaginary phase)**

---

### 3. p3 = '-iZX'

$$
p3 = -i \cdot (Z \otimes X)
$$

- Z on first qubit, X on second
- Multiplied by **-i**

---

## 🔹 General Concept (Pauli Strings)

A Pauli operator is:

$$
P = \alpha (P_1 \otimes P_2 \otimes \cdots \otimes P_n)
$$

Where:
- $P_i \in \{I, X, Y, Z\}$
- $\alpha \in \{1, -1, i, -i\}$

---

## 🔹 Key Ideas

### 1. These are Operators (NOT states)

They act on quantum states:

$$
P|\psi\rangle
$$

---

### 2. Tensor Product Structure

- Each letter = operation on one qubit
- Whole string = combined operation

Example:

$$
XZ = X \otimes Z
$$

---

### 3. Phase Factors Matter

- 'i' and '-i' are **global phase factors**
- They arise from Pauli multiplication rules:

$$
XY = iZ, \quad YX = -iZ
$$

---

## 🔹 Action Example

On a 2-qubit state:

$$
(X \otimes Z)|00\rangle = |10\rangle
$$

- X flips first qubit
- Z keeps second qubit unchanged (since it's |0⟩)

---

## 🔹 What Print Shows

When printed:

XZ, iYZ, -iZX

This is:
- **symbolic representation**
- not the matrix form

---

## 🔹 Why Important

These operators are used in:

- Quantum Hamiltonians  
- Observables (e.g., $Z \otimes Z$)  
- Variational algorithms (VQE, QAOA)  
- Quantum machine learning  
- Error correction  

---

## 🔹 Key Takeaways

- Pauli strings = multi-qubit operations  
- Each character acts on one qubit  
- Prefix (i, -i) = phase factor  
- Fundamental building blocks in quantum computing

# Pauli Representation using (z, x) Arrays — Notes

Given:

z = [False, False, True, True]  
x = [False, True, False, True]  

Pauli((z, x))

---

## 🔹 What is this representation?

This is the **binary (symplectic) representation** of a Pauli operator.

Each qubit is described using two bits:

- z[i] → controls Z component  
- x[i] → controls X component  

---

## 🔹 Mapping Rule (Very Important)

For each qubit i:

| z | x | Operator |
|---|---|----------|
| 0 | 0 | I        |
| 0 | 1 | X        |
| 1 | 0 | Z        |
| 1 | 1 | Y        |

(Note: Y = iXZ, phase handled separately)

---

## 🔹 Apply Mapping to Given Arrays

z = [0, 0, 1, 1]  
x = [0, 1, 0, 1]  

Now evaluate each position:

### Qubit-wise:

1. (0, 0) → I  
2. (0, 1) → X  
3. (1, 0) → Z  
4. (1, 1) → Y  

---

## 🔹 Final Pauli String

$$
P = I \otimes X \otimes Z \otimes Y
$$

So output:

IXZY



## 🔹 How to Read

- Pair values **index-wise**: (z[i], x[i])  
- Each pair → operator on that qubit  
- NOT row-wise, NOT separate arrays  

---

## 🔹 Example

z = [0, 0, 1, 1]  
x = [0, 1, 0, 1]  

Pairs:

- (0,0) → I  
- (0,1) → X  
- (1,0) → Z  
- (1,1) → Y  

---

## 🔹 Final Operator

$$
I \otimes X \otimes Z \otimes Y
$$

→ **IXZY**

---

## 🔹 Key Point

Each qubit is described by **2 bits (z, x)** → determines I, X, Y, or Z.

# ScalarOp in Qiskit

`ScalarOp` in Qiskit represents a scalar multiple of the identity operator.

Mathematically:

$$
aI
$$

where:

- $a$ = scalar value (`coeff`)
- $I$ = identity matrix

---

# Basic Syntax

```python
ScalarOp(dims, coeff=1)
```

## Arguments

| Argument | Meaning |
|---|---|
| `dims` | Defines subsystem dimensions |
| `coeff` | Scalar multiplying the identity operator |

---

# Identity Matrix

For 1 qubit, the identity matrix is:

$$
I_2 =
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
$$

If we multiply it by 3:

$$
3I_2 =
3
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
=
\begin{bmatrix}
3 & 0 \\
0 & 3
\end{bmatrix}
$$

This is what `ScalarOp` creates.

---

# Example 1 : Single Qubit

```python
from qiskit.quantum_info import ScalarOp

op = ScalarOp(2, coeff=3)

print(op)
print(op.to_matrix())
```

Output matrix:

$$
\begin{bmatrix}
3 & 0 \\
0 & 3
\end{bmatrix}
$$

Explanation:

- `dims = 2`
- dimension 2 corresponds to one qubit
- identity matrix size becomes $2 \times 2$

---

# Understanding `dims`

`dims` specifies subsystem dimensions.

Examples:

| dims | Meaning |
|---|---|
| `2` | one qubit |
| `(2,2)` | two qubits |
| `(2,2,2)` | three qubits |
| `3` | one qutrit |
| `(2,3)` | qubit + qutrit |

---

# Example 2 : Two Qubits

```python
op = ScalarOp((2,2), coeff=5)
```

Here:

$$
2 \times 2 = 4
$$

So total Hilbert-space dimension is 4.

The operator becomes:

$$
5I_4
$$

Matrix form:

$$
\begin{bmatrix}
5 & 0 & 0 & 0 \\
0 & 5 & 0 & 0 \\
0 & 0 & 5 & 0 \\
0 & 0 & 0 & 5
\end{bmatrix}
$$

---

# Example 3 : Four Qubits

```python
op = ScalarOp((2,2,2,2), coeff=7)
```

This means:

- 4 subsystems
- each subsystem has dimension 2
- therefore 4 qubits

Total dimension:

$$
2 \times 2 \times 2 \times 2 = 16
$$

So the operator becomes:

$$
7I_{16}
$$

which is a $16 \times 16$ identity matrix multiplied by 7.

---

# Why does Qiskit use subsystem dimensions?

In quantum computing, tensor-product structure matters.

Example:

```python
dims=(2,2,2)
```

means:

- qubit 1
- qubit 2
- qubit 3

instead of simply dimension 8.

Qiskit stores subsystem information internally because quantum operations act separately on qubits.

---

# Important Concept

`ScalarOp` does NOT create arbitrary matrices.

It ONLY creates:

$$
\text{scalar} \times \text{identity matrix}
$$

So:

```python
ScalarOp((2,2), coeff=4)
```

creates:

$$
4I_4
$$

NOT a random $4 \times 4$ matrix.

---

# Simple Mental Model

```python
ScalarOp(dims, coeff)
```

means:

> Create an identity matrix whose size is determined by `dims`,
> then multiply every diagonal element by `coeff`.

# Pauli `z` and `x` Arrays in Qiskit

In Qiskit, Pauli operators are internally stored using two binary arrays:

- `z` array
- `x` array

These arrays are a compact binary encoding of Pauli gates.

---

# Main Idea

Instead of storing full matrices, Qiskit stores simple binary information.

Each qubit gets:

- one `z` bit
- one `x` bit

Qiskit later combines these bits to reconstruct the Pauli operator.

---

# Binary Encoding Rule

| Pauli Gate | z bit | x bit |
|---|---|---|
| I | 0 | 0 |
| X | 0 | 1 |
| Z | 1 | 0 |
| Y | 1 | 1 |

---

# Why does Y have both bits = 1?

Because:

$$
Y = iXZ
$$

So Y contains:

- X component
- Z component

Therefore:

$$
z = 1,\quad x = 1
$$

---

# Example 1

```python
from qiskit.quantum_info import Pauli

m = Pauli("XYZ")

print(m.z)
print(m.x)
```

Output:

```python
[False  True  True]
[ True  True False]
```

Boolean values mean:

| Boolean | Binary |
|---|---|
| False | 0 |
| True | 1 |

So:

```python
z = [0 1 1]
x = [1 1 0]
```

---

# Understanding the Arrays

Read corresponding positions together.

---

## Qubit 1

$$
z=0,\quad x=1
$$

Using the encoding table:

| z | x | Gate |
|---|---|---|
| 0 | 1 | X |

So qubit 1 = X

---

## Qubit 2

$$
z=1,\quad x=1
$$

| z | x | Gate |
|---|---|---|
| 1 | 1 | Y |

So qubit 2 = Y

---

## Qubit 3

$$
z=1,\quad x=0
$$

| z | x | Gate |
|---|---|---|
| 1 | 0 | Z |

So qubit 3 = Z

---

# Final Reconstruction

Combining all qubits:

$$
XYZ
$$

So:

```python
z=[0,1,1]
x=[1,1,0]
```

represents:

```python
Pauli("XYZ")
```

---

# Important Concept

The `z` array alone does NOT represent the Pauli operator.

The `x` array alone also does NOT represent the operator.

Both arrays must be read together.

---

# Visual Interpretation

Think of each qubit having two switches:

- X switch
- Z switch

| X Switch | Z Switch | Result |
|---|---|---|
| OFF | OFF | I |
| ON | OFF | X |
| OFF | ON | Z |
| ON | ON | Y |

---

# Example 2

```python
m = Pauli("IXYZ")
```

| Qubit | Gate | z | x |
|---|---|---|---|
| 1 | I | 0 | 0 |
| 2 | X | 0 | 1 |
| 3 | Y | 1 | 1 |
| 4 | Z | 1 | 0 |

Arrays become:

```python
z = [0 0 1 1]
x = [0 1 1 0]
```

---

# Why does Qiskit use this representation?

Binary arrays are:

- memory efficient
- faster for computation
- useful for stabilizer circuits
- important in quantum error correction

instead of storing large matrices.

---

# Simple Mental Model

`z` array:

> Which qubits contain Z behavior?

`x` array:

> Which qubits contain X behavior?

Combining both arrays reconstructs the Pauli operator.
